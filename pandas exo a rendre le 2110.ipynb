```python
#afficher les femmes qui ont plus de 50 ans
df[(df['age']>50) & (df['sexe']=='feminin')]
```

```python
#verifier s'il y a des donnees manquantes
df.isnull().sum()
```

```python
#suppression les lignes ou il y a des donnees maquantes
df.dropna()
```

```python
#Recodage de sexe (0 pour masculin et 1 pour féminin)
df['sexe']=df['sexe'].replace(['feminin','masculin'],[1,0])
df
    
```

```python
#supprimer les variable depression, vaisseau et pic du dataframe
df.drop(columns=['depression','vaisseau','pic'],inplace=True)
```

```python
#Representation graphique de la variable coeur
df['coeur']
df.plot('coeur',kind='bar')
```

```python
#Representation graphique de toutes les variables qualitatives
#creons une listequal
listequal=df.select_dtypes(include='object').columns
listequal
plt.figure(figsize=(10,51))
a=1
for alphaa in listequal:
    plt.subplot(1,4,a)
    df[alphaa].value_counts().plot(kind='pie')
    plt.legend(alphaa)
    a+=1
    plt.show()
```

```python
#Representation graphique de toutes les variables quantitative
listequant=df.select_dtypes(exclude='object').columns
listequant
plt.figure(figsize=(10,21))
b=1
for omega in listequant:
    plt.subplot(4,4,b)
    df[omega].value_counts().plot(kind='hist',color='gray')
    b+=1
    plt.show()
```
